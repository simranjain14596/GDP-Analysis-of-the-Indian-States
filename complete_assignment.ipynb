## Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4, where C1 would have the highest per capita GDP and C4, the lowest).

```python
#defining the quantile
df3 = df2[:]['Per Capita GSDP (Rs.)']
m = df3.quantile(.20)
n = df3.quantile(.50)
o = df3.quantile(.85)
p = df3.quantile(1)
```

```python
#categorizing the data inti c1,c2,c3 ,c4 based on gsdp
df2['category']=''
for ind in df2.index:  
    i = df2['Per Capita GSDP (Rs.)'][ind]
    if i<= m:
        df2['category'][ind] = 'c4'
    elif i<=n:
        df2['category'][ind] = 'c3'
    elif i<=o:
        df2['category'][ind] = 'c2'
    else:
        df2['category'][ind] = 'c1'
```

```python
#defining the pivot table with index as category
datafr=df2.pivot_table(values = ['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing','Electricity, gas, water supply & other utility services','Construction','Trade, repair, hotels and restaurants','Transport, storage, communication & services related to broadcasting','Financial services','Real estate, ownership of dwelling & professional services','Public administration','Other services','Gross State Domestic Product'], index = 'category', aggfunc = 'sum')
```

```python
datafr.head(1)
```

```python
my_list=datafr.columns
```

```python
#filtering the data as data has column "GSDP"
for i in  my_list:
    if i == 'Gross State Domestic Product':
        continue
    else:
        for j in datafr.index:
            datafr[i][j]=round(((datafr[i][j]/datafr['Gross State Domestic Product'][j])*100),1)
```

```python
datafr.info()
datafr.drop(['Gross State Domestic Product'],axis=1,inplace = True)
```

```python
#transpose the dataframe so that you get the exact graph
datafr = datafr.T
```

```python
datafr.head(5)
```

```python
#sort the values of the dataframe
c1=datafr[['c1']]
c1.reset_index(inplace=True)
c1.rename(columns={"index": "Sector", "c1": "percent_of_GDP"},inplace=True)
c1.sort_values(by=['percent_of_GDP'],inplace=True,ascending=False)
c2=datafr[['c2']]
c2.reset_index(inplace=True)
c2.rename(columns={"index": "Sector", "c2": "percent_of_GDP"},inplace=True)
c2.sort_values(by=['percent_of_GDP'],inplace=True,ascending=False)
c3=datafr[['c3']]
c3.reset_index(inplace=True)
c3.rename(columns={"index": "Sector", "c3": "percent_of_GDP"},inplace=True)
c3.sort_values(by=['percent_of_GDP'],inplace=True,ascending=False)
c4=datafr[['c4']]
c4.reset_index(inplace=True)
c4.rename(columns={"index": "Sector", "c4": "percent_of_GDP"},inplace=True)
c4.sort_values(by=['percent_of_GDP'],inplace=True,ascending=False)
```
